```python
data = pd.read_csv('IMDB Dataset.csv')
data.head(5)
```

```python
data.isnull().sum()
```

```python
plt.hist(data['Runtime'])
plt.ylabel("Count of movies")
plt.xlabel("Runtime")
```

```python
plt.hist(data['IMDB_Rating'])
plt.ylabel("Count pf movies")
plt.xlabel("IMDB Rating")
```

```python
data['Runtime'].unique()

bins_time = np.arange(80, 230, 10)
plt.hist(data['Runtime'], bins=bins_time, edgecolor='black', color='g')
plt.ylabel("Count of movies")
plt.xlabel("Runtime")
```

```python
data['IMDB_Rating'].unique()

bins_rating = np.arange(8, 10, 0.20)
plt.hist(data['IMDB_Rating'], bins=bins_rating, edgecolor='black', color='g')
plt.ylabel("Count of movies")
plt.xlabel("IMDB Rating")
plt.xticks(bins_rating)
```
